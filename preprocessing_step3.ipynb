```python
import pandas as pd
import numpy as np
import os

pd.options.display.max_columns = None
```

```python
path = 'user_bids_dfs_testing/'
# path = 'user_bids_dfs_training/'
files = os.listdir(path)

count = 0
bidder_dictinoary = {}

for file in files:
    
    count += 1
    
    user_id = file.split(".")[0]
    
    bidder_dictinoary[user_id] = pd.read_csv(path + file, index_col=None)
    print('counter: {}, fileloaded: {}, df_shape:{}'.format(count, file, bidder_dictinoary[user_id].shape))
```

```python
list_of_df = []

i = 0

for bidder in bidder_dictinoary:
    
    df = bidder_dictinoary[bidder]
    
      # add device data
    device_values = df['device'].value_counts()

    df['bids_on_devices'] = 0

    for device in df['device'].unique():
        df.loc[df['device']==device, 'bids_on_devices'] = device_values[device]
        
    device_values_described = df['device'].value_counts().describe()

    describe_axes = device_values_described.axes[0]

    for i in range(len(describe_axes)):
        df['{}_{}'.format('device_bids', describe_axes[i])] = device_values_described[i]
        
    df['date_time'] = pd.to_datetime(df['time'])
    df = df.sort_values(['date_time']) 
      
    # add time data for auctions in user frame
    time_delta_auction_bids_described = df['time_delta_auction_bids'].value_counts().describe()

    describe_axes = time_delta_auction_bids_described.axes[0]

    for i in range(1, len(describe_axes)):
        df['{}_{}'.format('time_delta_auction_user_bids', describe_axes[i])] = device_values_described[i]

    df['time_delta_auction_user_bids'] = df[1:]['time_delta_auction_bids'].median()
    df['time_delta_auction_user_bids'] = df[1:]['time_delta_auction_bids'].kurtosis()
    
    bid_counts_first = df['first_auction_bid'].value_counts()
    bid_counts_wins = df['last_auction_bid'].value_counts()
    
    bid_counts_first = df['first_auction_bid'].value_counts()
    bid_counts_wins = df['last_auction_bid'].value_counts()
    
    df['first_bid_over_total_bids'] = 0
    df['win_bid_over_total_bids'] = 0
    
    if 1 in bid_counts_first:
        first_bid_over_total_bids = bid_counts_first[1] / df.shape[0]
        df['first_bid_over_total_bids'] = first_bid_over_total_bids
        
    if 1 in bid_counts_wins:
        win_bid_over_total_bids = bid_counts_wins[1] / df.shape[0]
        df['win_bid_over_total_bids'] = win_bid_over_total_bids
        
    df = df.drop_duplicates(subset='bidder_id', keep="first")
    
    list_of_df.append(df)

unordered_df = pd.concat(list_of_df, ignore_index=True)
```
