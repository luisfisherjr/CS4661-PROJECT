```python
biderID = df_joined_train['bidder_id'].unique()
```

```python
biderID_df_dictionary = {}

for bidder_id in biderID:
    biderID_df_dictionary[bidder_id] = df_joined_train.loc[df_joined_train['bidder_id'] == bidder_id]
    
    print('id: {}, df_shape: {}'.format(bidder_id, biderID_df_dictionary[bidder_id].shape))
```

```python
i = 0

bot_ip = set(bots['ip'].unique())
p0 = None

for bidder_id in biderID:
    i = i + 1
    
    if i % 50 == 0:
        print(i)
        
    user_ip = set(biderID_df_dictionary[bidder_id]['ip'].unique()) 
    
    tdelta = biderID_df_dictionary[bidder_id]['time'].diff()
    tdelta.fillna(0 , inplace=True)
    biderID_df_dictionary[bidder_id]['time_between_bids'] = pd.to_timedelta(tdelta) / np.timedelta64(1, 's')
    
    biderID_df_dictionary[bidder_id]['total_bids'] = biderID_df_dictionary[bidder_id]['time_between_bids'].size
    
    biderID_df_dictionary[bidder_id]['time_between_bids_median'] = \
    biderID_df_dictionary[bidder_id]['time_between_bids'].median()
    
    biderID_df_dictionary[bidder_id]['time_between_bids_mean'] = \
    biderID_df_dictionary[bidder_id]['time_between_bids'].mean()
    
    biderID_df_dictionary[bidder_id]['avg_bids_per_url'] =  \
    biderID_df_dictionary[bidder_id]['time_between_bids'].size / len(biderID_df_dictionary[bidder_id]['url'].unique())
    
    biderID_df_dictionary[bidder_id]['fraction_of_bot_shared_ips'] = len(bot_ip & user_ip) / len(user_ip)
    
    biderID_df_dictionary[bidder_id].to_csv('user_bids_dfs/{}.csv'.format(bidder_id))
    
```
